```
# –ò—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
text = """
Python —è–≤–ª—è–µ—Ç—Å—è –æ–¥–Ω–∏–º –∏–∑ —Å–∞–º—ã—Ö –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö —è–∑—ã–∫–æ–≤ –ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞–Ω–∏—è.
–Ø–∑—ã–∫ Python –ø—Ä–æ—Å—Ç –≤ –∏–∑—É—á–µ–Ω–∏–∏ –∏ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–µ–Ω –≤ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–∏.
Python –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –≤ –≤–µ–±-—Ä–∞–∑—Ä–∞–±–æ—Ç–∫–µ, –∞–Ω–∞–ª–∏–∑–µ –¥–∞–Ω–Ω—ã—Ö, –∏—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω–æ–º –∏–Ω—Ç–µ–ª–ª–µ–∫—Ç–µ.
"""

def analyze_text(text):
    """
    –ü—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å —Ç–µ–∫—Å—Ç –∏ –≤–µ—Ä–Ω—É—Ç—å —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É:
    - –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–ª–æ–≤
    - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Å–ª–æ–≤
    - —Å–ø–∏—Å–æ–∫ —Å–ª–æ–≤ –≤ –≤–µ—Ä—Ö–Ω–µ–º —Ä–µ–≥–∏—Å—Ç—Ä–µ
    """
    # –†–∞–∑–±–∏—Ç—å —Ç–µ–∫—Å—Ç –Ω–∞ —Å–ª–æ–≤–∞ (—É–±—Ä–∞—Ç—å –∑–Ω–∞–∫–∏ –ø—Ä–µ–ø–∏–Ω–∞–Ω–∏—è –∏ –ø—Ä–∏–≤–µ—Å—Ç–∏ –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É)
    words = text.lower().split()
    cleaned_words = [word.strip('.,!?;:()') for word in words]

    # TODO: –í—ã—á–∏—Å–ª–∏—Ç—å –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–ª–æ–≤
    total_words = len(cleaned_words)

    # TODO: –ù–∞–π—Ç–∏ —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ —Å–ª–æ–≤–∞ —Å –ø–æ–º–æ—â—å—é –º–Ω–æ–∂–µ—Å—Ç–≤–∞
    unique_words = set(cleaned_words)

    # TODO: –°–æ–∑–¥–∞—Ç—å —Å–ø–∏—Å–æ–∫ —Å–ª–æ–≤ –≤ –≤–µ—Ä—Ö–Ω–µ–º —Ä–µ–≥–∏—Å—Ç—Ä–µ (–¥–ª–∏–Ω–Ω–µ–µ 3 —Å–∏–º–≤–æ–ª–æ–≤)
    uppercase_words = [word.upper() for word in cleaned_words if len(word) > 3]

    return {
        "total_words": total_words,
        "unique_words_count": len(unique_words),
        "unique_words": unique_words,
        "uppercase_words": uppercase_words
    }

# –ü—Ä–æ–≤–µ—Ä–∫–∞ —Ä–∞–±–æ—Ç—ã —Ñ—É–Ω–∫—Ü–∏–∏
result = analyze_text(text)
print(f"–û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–ª–æ–≤: {result['total_words']}")
print(f"–£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Å–ª–æ–≤: {result['unique_words_count']}")
print(f"–ü—Ä–∏–º–µ—Ä —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Å–ª–æ–≤: {list(result['unique_words'])[:5]}")
```

```
def create_student_database():
    """
    –°–æ–∑–¥–∞—Ç—å –±–∞–∑—É –¥–∞–Ω–Ω—ã—Ö —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ –∏ —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –æ–ø–µ—Ä–∞—Ü–∏–∏:
    - –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ —Å—Ç—É–¥–µ–Ω—Ç–∞
    - –ø–æ–∏—Å–∫ –ø–æ –∏–º–µ–Ω–∏
    - —Ä–∞—Å—á–µ—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ –±–∞–ª–ª–∞
    - –≤—ã–≤–æ–¥ —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ —Å –æ—Ü–µ–Ω–∫–æ–π –≤—ã—à–µ –∑–∞–¥–∞–Ω–Ω–æ–π
    """
    students = {}

    def add_student(student_id, name, age, grades):
        """
        –î–æ–±–∞–≤–∏—Ç—å —Å—Ç—É–¥–µ–Ω—Ç–∞ –≤ –±–∞–∑—É –¥–∞–Ω–Ω—ã—Ö
        grades –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å —Å–ø–∏—Å–∫–æ–º –æ—Ü–µ–Ω–æ–∫
        """
        # TODO: –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ —Å—Ç—É–¥–µ–Ω—Ç–∞
        students[student_id] = {
            "–∏–º—è": name,
            "–≤–æ–∑—Ä–∞—Å—Ç": age,
            "–æ—Ü–µ–Ω–∫–∏": grades,
            "—Å—Ä–µ–¥–Ω–∏–π_–±–∞–ª–ª": sum(grades) / len(grades) if grades else 0
        }

    def find_student(name):
        """–ù–∞–π—Ç–∏ –≤—Å–µ—Ö —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ —Å —É–∫–∞–∑–∞–Ω–Ω—ã–º –∏–º–µ–Ω–µ–º"""
        # TODO: –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –ø–æ–∏—Å–∫ –ø–æ –∏–º–µ–Ω–∏
        found = []
        for student_id, info in students.items():
            if info["–∏–º—è"].lower() == name.lower():
                found.append((student_id, info))
        return found

    def get_top_students(threshold=4.0):
        """–í–µ—Ä–Ω—É—Ç—å —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ —Å–æ —Å—Ä–µ–¥–Ω–∏–º –±–∞–ª–ª–æ–º –≤—ã—à–µ threshold"""
        # TODO: –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å —Ñ–∏–ª—å—Ç—Ä–∞—Ü–∏—é –ø–æ —Å—Ä–µ–¥–Ω–µ–º—É –±–∞–ª–ª—É
        top_students = {}
        for student_id, info in students.items():
            if info["—Å—Ä–µ–¥–Ω–∏–π_–±–∞–ª–ª"] >= threshold:
                top_students[student_id] = info
        return top_students

    # –î–æ–±–∞–≤–ª—è–µ–º —Ç–µ—Å—Ç–æ–≤—ã—Ö —Å—Ç—É–¥–µ–Ω—Ç–æ–≤
    add_student(1, "–ê–Ω–Ω–∞ –ò–≤–∞–Ω–æ–≤–∞", 20, [5, 4, 5, 5])
    add_student(2, "–ü–µ—Ç—Ä –°–∏–¥–æ—Ä–æ–≤", 21, [3, 4, 4, 3])
    add_student(3, "–ú–∞—Ä–∏—è –ö–æ–∑–ª–æ–≤–∞", 19, [5, 5, 5, 5])

    # TODO: –ü—Ä–æ—Ç–µ—Å—Ç–∏—Ä–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏–∏
    print("–í—Å–µ —Å—Ç—É–¥–µ–Ω—Ç—ã:", students)
    print("–ü–æ–∏—Å–∫ '–ê–Ω–Ω–∞':", find_student("–ê–Ω–Ω–∞"))
    print("–¢–æ–ø —Å—Ç—É–¥–µ–Ω—Ç—ã:", get_top_students(4.5))

    return students

# –ó–∞–ø—É—Å–∫ –±–∞–∑—ã –¥–∞–Ω–Ω—ã—Ö
database = create_student_database()
```

```
def analyze_sales_data():
    """
    –ü—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å –¥–∞–Ω–Ω—ã–µ –æ –ø—Ä–æ–¥–∞–∂–∞—Ö –∏ –≤—ã–≤–µ—Å—Ç–∏:
    - –æ–±—â—É—é –≤—ã—Ä—É—á–∫—É –ø–æ –∫–∞–∂–¥–æ–º—É —Ç–æ–≤–∞—Ä—É
    - —Å–∞–º—ã–π –ø–æ–ø—É–ª—è—Ä–Ω—ã–π —Ç–æ–≤–∞—Ä
    - —Å—Ä–µ–¥–Ω—é—é —Å—Ç–æ–∏–º–æ—Å—Ç—å –∑–∞–∫–∞–∑–∞
    """

    # –î–∞–Ω–Ω—ã–µ –æ –ø—Ä–æ–¥–∞–∂–∞—Ö: —Å–ø–∏—Å–æ–∫ —Å–ª–æ–≤–∞—Ä–µ–π
    sales = [
        {"product": "–Ω–æ—É—Ç–±—É–∫", "price": 50000, "quantity": 2, "date": "2024-01-15"},
        {"product": "–º—ã—à—å", "price": 1500, "quantity": 5, "date": "2024-01-15"},
        {"product": "–Ω–æ—É—Ç–±—É–∫", "price": 50000, "quantity": 1, "date": "2024-01-16"},
        {"product": "–∫–ª–∞–≤–∏–∞—Ç—É—Ä–∞", "price": 3000, "quantity": 3, "date": "2024-01-16"},
        {"product": "–º–æ–Ω–∏—Ç–æ—Ä", "price": 20000, "quantity": 2, "date": "2024-01-17"},
    ]

    def calculate_product_revenue(sales_data):
        """–†–∞—Å—Å—á–∏—Ç–∞—Ç—å –æ–±—â—É—é –≤—ã—Ä—É—á–∫—É –ø–æ –∫–∞–∂–¥–æ–º—É —Ç–æ–≤–∞—Ä—É"""
        revenue_by_product = {}
        for sale in sales_data:
            product = sale["product"]
            revenue = sale["price"] * sale["quantity"]

            if product in revenue_by_product:
                revenue_by_product[product] += revenue
            else:
                revenue_by_product[product] = revenue
        return revenue_by_product

    def find_most_popular_product(sales_data):
        """–ù–∞–π—Ç–∏ —Ç–æ–≤–∞—Ä —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –ø—Ä–æ–¥–∞–∂"""
        quantity_by_product = {}
        for sale in sales_data:
            product = sale["product"]
            quantity = sale["quantity"]

            if product in quantity_by_product:
                quantity_by_product[product] += quantity
            else:
                quantity_by_product[product] = quantity

        # TODO: –ù–∞–π—Ç–∏ —Ç–æ–≤–∞—Ä —Å –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º
        if quantity_by_product:
            return max(quantity_by_product, key=quantity_by_product.get)
        return None

    def calculate_average_order_value(sales_data):
        """–†–∞—Å—Å—á–∏—Ç–∞—Ç—å —Å—Ä–µ–¥–Ω—é—é —Å—Ç–æ–∏–º–æ—Å—Ç—å –æ–¥–Ω–æ–≥–æ –∑–∞–∫–∞–∑–∞"""
        total_revenue = 0
        total_orders = len(sales_data)

        for sale in sales_data:
            total_revenue += sale["price"] * sale["quantity"]

        return total_revenue / total_orders if total_orders > 0 else 0

    # –í—ã–ø–æ–ª–Ω—è–µ–º –∞–Ω–∞–ª–∏–∑
    product_revenue = calculate_product_revenue(sales)
    popular_product = find_most_popular_product(sales)
    avg_order = calculate_average_order_value(sales)

    print("–í—ã—Ä—É—á–∫–∞ –ø–æ —Ç–æ–≤–∞—Ä–∞–º:", product_revenue)
    print("–°–∞–º—ã–π –ø–æ–ø—É–ª—è—Ä–Ω—ã–π —Ç–æ–≤–∞—Ä:", popular_product)
    print("–°—Ä–µ–¥–Ω—è—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –∑–∞–∫–∞–∑–∞:", f"{avg_order:.2f} —Ä—É–±.")

    return product_revenue

# –ó–∞–ø—É—Å–∫ –∞–Ω–∞–ª–∏–∑–∞
analysis_results = analyze_sales_data()
```

```
# @title –°–∏—Å—Ç–µ–º–∞ –æ–±—Ä–∞–±–æ—Ç–∫–∏ –∑–∞–∫–∞–∑–æ–≤ –¥–ª—è –∫–∞—Ñ–µ
import pandas as pd
from datetime import datetime
from google.colab import files
import ipywidgets as widgets
from IPython.display import display, clear_output
import matplotlib.pyplot as plt

class CafeOrderSystem:
    def __init__(self):
        self.menu = pd.DataFrame(columns=['id', '–Ω–∞–∑–≤–∞–Ω–∏–µ', '—Ü–µ–Ω–∞', '–∫–∞—Ç–µ–≥–æ—Ä–∏—è', '–≤—Ä–µ–º—è_–ø—Ä–∏–≥–æ—Ç–æ–≤–ª–µ–Ω–∏—è'])
        self.orders = pd.DataFrame(columns=['id', '—Å—Ç–æ–ª', '–±–ª—é–¥–∞', '—Å—Ç–∞—Ç—É—Å', '—Å—É–º–º–∞', '–¥–∞—Ç–∞_–≤—Ä–µ–º—è'])
        self.next_menu_id = 1
        self.next_order_id = 1

    def add_dish(self, name, price, category, prep_time):
        """–î–æ–±–∞–≤–ª–µ–Ω–∏–µ –±–ª—é–¥–∞ –≤ –º–µ–Ω—é"""
        new_dish = pd.DataFrame({
            'id': [self.next_menu_id],
            '–Ω–∞–∑–≤–∞–Ω–∏–µ': [name],
            '—Ü–µ–Ω–∞': [price],
            '–∫–∞—Ç–µ–≥–æ—Ä–∏—è': [category],
            '–≤—Ä–µ–º—è_–ø—Ä–∏–≥–æ—Ç–æ–≤–ª–µ–Ω–∏—è': [prep_time]
        })
        self.menu = pd.concat([self.menu, new_dish], ignore_index=True)
        self.next_menu_id += 1
        print(f"‚úÖ –ë–ª—é–¥–æ '{name}' –¥–æ–±–∞–≤–ª–µ–Ω–æ –≤ –º–µ–Ω—é!")

    def create_order(self, table, dish_ids):
        """–°–æ–∑–¥–∞–Ω–∏–µ –∑–∞–∫–∞–∑–∞"""
        # –ü–æ–ª—É—á–∞–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –∑–∞–∫–∞–∑–∞–Ω–Ω—ã—Ö –±–ª—é–¥–∞—Ö
        ordered_dishes = self.menu[self.menu['id'].isin(dish_ids)]

        if ordered_dishes.empty:
            print("‚ùå –û—à–∏–±–∫–∞: —É–∫–∞–∑–∞–Ω—ã –Ω–µ–≤–µ—Ä–Ω—ã–µ ID –±–ª—é–¥!")
            return

        # –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º —Å—É–º–º—É –∑–∞–∫–∞–∑–∞
        total = ordered_dishes['—Ü–µ–Ω–∞'].sum()

        # –°–æ–∑–¥–∞–µ–º –∑–∞–ø–∏—Å—å –æ –∑–∞–∫–∞–∑–µ
        dishes_list = ordered_dishes['–Ω–∞–∑–≤–∞–Ω–∏–µ'].tolist()

        new_order = pd.DataFrame({
            'id': [self.next_order_id],
            '—Å—Ç–æ–ª': [table],
            '–±–ª—é–¥–∞': [', '.join(dishes_list)],
            '—Å—Ç–∞—Ç—É—Å': ['–ø—Ä–∏–Ω—è—Ç'],
            '—Å—É–º–º–∞': [total],
            '–¥–∞—Ç–∞_–≤—Ä–µ–º—è': [datetime.now().strftime("%Y-%m-%d %H:%M:%S")]
        })

        self.orders = pd.concat([self.orders, new_order], ignore_index=True)
        self.next_order_id += 1

        print(f"‚úÖ –ó–∞–∫–∞–∑ #{self.next_order_id-1} —Å–æ–∑–¥–∞–Ω!")
        print(f"   –°—Ç–æ–ª: {table}")
        print(f"   –ë–ª—é–¥–∞: {', '.join(dishes_list)}")
        print(f"   –°—É–º–º–∞: {total} —Ä—É–±.")

    def update_order_status(self, order_id, new_status):
        """–û–±–Ω–æ–≤–ª–µ–Ω–∏–µ —Å—Ç–∞—Ç—É—Å–∞ –∑–∞–∫–∞–∑–∞"""
        valid_statuses = ['–ø—Ä–∏–Ω—è—Ç', '–≥–æ—Ç–æ–≤–∏—Ç—Å—è', '–≥–æ—Ç–æ–≤', '–æ–ø–ª–∞—á–µ–Ω', '–æ—Ç–º–µ–Ω–µ–Ω']

        if new_status not in valid_statuses:
            print(f"‚ùå –ù–µ–≤–µ—Ä–Ω—ã–π —Å—Ç–∞—Ç—É—Å! –î–æ–ø—É—Å—Ç–∏–º—ã–µ —Å—Ç–∞—Ç—É—Å—ã: {', '.join(valid_statuses)}")
            return

        if order_id in self.orders['id'].values:
            self.orders.loc[self.orders['id'] == order_id, '—Å—Ç–∞—Ç—É—Å'] = new_status
            print(f"‚úÖ –°—Ç–∞—Ç—É—Å –∑–∞–∫–∞–∑–∞ #{order_id} –∏–∑–º–µ–Ω–µ–Ω –Ω–∞ '{new_status}'")
        else:
            print(f"‚ùå –ó–∞–∫–∞–∑ #{order_id} –Ω–µ –Ω–∞–π–¥–µ–Ω!")

    def calculate_order_cost(self, order_id):
        """–†–∞—Å—á–µ—Ç —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –∑–∞–∫–∞–∑–∞"""
        order = self.orders[self.orders['id'] == order_id]
        if not order.empty:
            print(f"üí∞ –°—Ç–æ–∏–º–æ—Å—Ç—å –∑–∞–∫–∞–∑–∞ #{order_id}: {order.iloc[0]['—Å—É–º–º–∞']} —Ä—É–±.")
            return order.iloc[0]['—Å—É–º–º–∞']
        else:
            print(f"‚ùå –ó–∞–∫–∞–∑ #{order_id} –Ω–µ –Ω–∞–π–¥–µ–Ω!")
            return 0

    def get_popular_dishes(self, n=5):
        """–ü–æ–∏—Å–∫ —Å–∞–º—ã—Ö –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –±–ª—é–¥"""
        all_dishes = []

        for dishes in self.orders['–±–ª—é–¥–∞']:
            all_dishes.extend([d.strip() for d in dishes.split(',')])

        if not all_dishes:
            print("üìä –ü–æ–∫–∞ –Ω–µ—Ç –¥–∞–Ω–Ω—ã—Ö –æ –∑–∞–∫–∞–∑–∞—Ö")
            return pd.DataFrame()

        dish_counts = pd.Series(all_dishes).value_counts().head(n)

        print(f"üìä –¢–û–ü-{n} —Å–∞–º—ã—Ö –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –±–ª—é–¥:")
        for i, (dish, count) in enumerate(dish_counts.items(), 1):
            print(f"   {i}. {dish} - {count} –∑–∞–∫–∞–∑(–æ–≤)")

        return dish_counts

    def revenue_by_category(self):
        """–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø–æ –≤—ã—Ä—É—á–∫–µ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º"""
        if self.orders.empty:
            print("üìä –ù–µ—Ç –¥–∞–Ω–Ω—ã—Ö –æ –∑–∞–∫–∞–∑–∞—Ö")
            return pd.DataFrame()

        # –ü–æ–ª—É—á–∞–µ–º –≤—Å–µ –∑–∞–∫–∞–∑—ã —Å–æ —Å—Ç–∞—Ç—É—Å–æ–º "–æ–ø–ª–∞—á–µ–Ω"
        paid_orders = self.orders[self.orders['—Å—Ç–∞—Ç—É—Å'] == '–æ–ø–ª–∞—á–µ–Ω']

        if paid_orders.empty:
            print("üìä –ù–µ—Ç –æ–ø–ª–∞—á–µ–Ω–Ω—ã—Ö –∑–∞–∫–∞–∑–æ–≤")
            return pd.DataFrame()

        # –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º –≤—ã—Ä—É—á–∫—É –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º
        revenue = {}

        for _, order in paid_orders.iterrows():
            dishes = [d.strip() for d in order['–±–ª—é–¥–∞'].split(',')]
            for dish in dishes:
                dish_info = self.menu[self.menu['–Ω–∞–∑–≤–∞–Ω–∏–µ'] == dish].iloc[0]
                category = dish_info['–∫–∞—Ç–µ–≥–æ—Ä–∏—è']
                price = dish_info['—Ü–µ–Ω–∞']

                if category in revenue:
                    revenue[category] += price
                else:
                    revenue[category] = price

        revenue_df = pd.DataFrame(list(revenue.items()),
                                 columns=['–ö–∞—Ç–µ–≥–æ—Ä–∏—è', '–í—ã—Ä—É—á–∫–∞'])
        revenue_df = revenue_df.sort_values('–í—ã—Ä—É—á–∫–∞', ascending=False)

        print("üìä –í—ã—Ä—É—á–∫–∞ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º:")
        for _, row in revenue_df.iterrows():
            print(f"   {row['–ö–∞—Ç–µ–≥–æ—Ä–∏—è']}: {row['–í—ã—Ä—É—á–∫–∞']} —Ä—É–±.")

        return revenue_df

    def show_menu(self):
        """–ü–æ–∫–∞–∑–∞—Ç—å –º–µ–Ω—é"""
        if self.menu.empty:
            print("üìã –ú–µ–Ω—é –ø—É—Å—Ç–æ")
        else:
            print("üìã –ú–ï–ù–Æ:")
            print(self.menu[['id', '–Ω–∞–∑–≤–∞–Ω–∏–µ', '—Ü–µ–Ω–∞', '–∫–∞—Ç–µ–≥–æ—Ä–∏—è', '–≤—Ä–µ–º—è_–ø—Ä–∏–≥–æ—Ç–æ–≤–ª–µ–Ω–∏—è']].to_string(index=False))

    def show_orders(self, status=None):
        """–ü–æ–∫–∞–∑–∞—Ç—å –∑–∞–∫–∞–∑—ã"""
        if self.orders.empty:
            print("üì¶ –ù–µ—Ç –∑–∞–∫–∞–∑–æ–≤")
            return

        display_orders = self.orders
        if status:
            display_orders = self.orders[self.orders['—Å—Ç–∞—Ç—É—Å'] == status]

        if display_orders.empty:
            print(f"üì¶ –ù–µ—Ç –∑–∞–∫–∞–∑–æ–≤ —Å–æ —Å—Ç–∞—Ç—É—Å–æ–º '{status}'")
        else:
            print(f"üì¶ –ó–ê–ö–ê–ó–´:" + (f" (—Å—Ç–∞—Ç—É—Å: {status})" if status else ""))
            print(display_orders[['id', '—Å—Ç–æ–ª', '–±–ª—é–¥–∞', '—Å—Ç–∞—Ç—É—Å', '—Å—É–º–º–∞', '–¥–∞—Ç–∞_–≤—Ä–µ–º—è']].to_string(index=False))

    def visualize_popular_dishes(self):
        """–í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –±–ª—é–¥"""
        popular = self.get_popular_dishes(5)
        if not popular.empty:
            plt.figure(figsize=(10, 6))
            popular.plot(kind='bar', color='skyblue')
            plt.title('–¢–û–ü-5 —Å–∞–º—ã—Ö –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –±–ª—é–¥')
            plt.xlabel('–ë–ª—é–¥–æ')
            plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤')
            plt.xticks(rotation=45, ha='right')
            plt.tight_layout()
            plt.show()

    def visualize_revenue_by_category(self):
        """–í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –≤—ã—Ä—É—á–∫–∏ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º"""
        revenue = self.revenue_by_category()
        if not revenue.empty:
            plt.figure(figsize=(10, 6))
            plt.pie(revenue['–í—ã—Ä—É—á–∫–∞'], labels=revenue['–ö–∞—Ç–µ–≥–æ—Ä–∏—è'], autopct='%1.1f%%')
            plt.title('–í—ã—Ä—É—á–∫–∞ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º')
            plt.axis('equal')
            plt.show()

# @title –ò–Ω—Ç–µ—Ä—Ñ–µ–π—Å –¥–ª—è —Ä–∞–±–æ—Ç—ã —Å —Å–∏—Å—Ç–µ–º–æ–π
# –°–æ–∑–¥–∞–µ–º —ç–∫–∑–µ–º–ø–ª—è—Ä —Å–∏—Å—Ç–µ–º—ã
cafe = CafeOrderSystem()

# –°–æ–∑–¥–∞–µ–º –≤–∏–¥–∂–µ—Ç—ã –¥–ª—è –∏–Ω—Ç–µ—Ä—Ñ–µ–π—Å–∞
print("=" * 50)
print("üè™ –°–ò–°–¢–ï–ú–ê –£–ü–†–ê–í–õ–ï–ù–ò–Ø –ó–ê–ö–ê–ó–ê–ú–ò –ö–ê–§–ï")
print("=" * 50)

# –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –±–ª—é–¥–∞
add_dish_btn = widgets.Button(description="‚ûï –î–æ–±–∞–≤–∏—Ç—å –±–ª—é–¥–æ", button_style='success')
dish_name = widgets.Text(placeholder="–ù–∞–∑–≤–∞–Ω–∏–µ –±–ª—é–¥–∞")
dish_price = widgets.FloatText(placeholder="–¶–µ–Ω–∞")
dish_category = widgets.Dropdown(options=['–ù–∞–ø–∏—Ç–∫–∏', '–ì–æ—Ä—è—á–µ–µ', '–°–∞–ª–∞—Ç—ã', '–î–µ—Å–µ—Ä—Ç—ã', '–ó–∞–∫—É—Å–∫–∏'],
                                 value='–ì–æ—Ä—è—á–µ–µ', description='–ö–∞—Ç–µ–≥–æ—Ä–∏—è:')
dish_time = widgets.IntText(placeholder="–í—Ä–µ–º—è –ø—Ä–∏–≥–æ—Ç–æ–≤–ª–µ–Ω–∏—è (–º–∏–Ω)")

# –°–æ–∑–¥–∞–Ω–∏–µ –∑–∞–∫–∞–∑–∞
create_order_btn = widgets.Button(description="üìù –°–æ–∑–¥–∞—Ç—å –∑–∞–∫–∞–∑", button_style='info')
order_table = widgets.IntText(placeholder="–ù–æ–º–µ—Ä —Å—Ç–æ–ª–∞")
order_dishes = widgets.Text(placeholder="ID –±–ª—é–¥ —á–µ—Ä–µ–∑ –∑–∞–ø—è—Ç—É—é (–Ω–∞–ø—Ä–∏–º–µ—Ä: 1,2,3)")

# –û–±–Ω–æ–≤–ª–µ–Ω–∏–µ —Å—Ç–∞—Ç—É—Å–∞
update_status_btn = widgets.Button(description="üîÑ –û–±–Ω–æ–≤–∏—Ç—å —Å—Ç–∞—Ç—É—Å", button_style='warning')
order_id_status = widgets.IntText(placeholder="ID –∑–∞–∫–∞–∑–∞")
new_status = widgets.Dropdown(options=['–ø—Ä–∏–Ω—è—Ç', '–≥–æ—Ç–æ–≤–∏—Ç—Å—è', '–≥–æ—Ç–æ–≤', '–æ–ø–ª–∞—á–µ–Ω', '–æ—Ç–º–µ–Ω–µ–Ω'],
                             value='–≥–æ—Ç–æ–≤–∏—Ç—Å—è', description='–ù–æ–≤—ã–π —Å—Ç–∞—Ç—É—Å:')

# –ö–Ω–æ–ø–∫–∏ –¥–ª—è –ø—Ä–æ—Å–º–æ—Ç—Ä–∞
show_menu_btn = widgets.Button(description="üìã –ü–æ–∫–∞–∑–∞—Ç—å –º–µ–Ω—é")
show_orders_btn = widgets.Button(description="üì¶ –ü–æ–∫–∞–∑–∞—Ç—å –∑–∞–∫–∞–∑—ã")
popular_btn = widgets.Button(description="üìä –ü–æ–ø—É–ª—è—Ä–Ω—ã–µ –±–ª—é–¥–∞")
revenue_btn = widgets.Button(description="üí∞ –í—ã—Ä—É—á–∫–∞ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º")
visualize_btn = widgets.Button(description="üìà –í–∏–∑—É–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É")

def on_add_dish_clicked(b):
    clear_output(wait=True)
    if dish_name.value and dish_price.value > 0:
        cafe.add_dish(dish_name.value, dish_price.value, dish_category.value, dish_time.value)
    else:
        print("‚ùå –ó–∞–ø–æ–ª–Ω–∏—Ç–µ –≤—Å–µ –ø–æ–ª—è –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ!")
    display_interface()

def on_create_order_clicked(b):
    clear_output(wait=True)
    try:
        dish_ids = [int(x.strip()) for x in order_dishes.value.split(',')]
        cafe.create_order(order_table.value, dish_ids)
    except:
        print("‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ —Å–æ–∑–¥–∞–Ω–∏–∏ –∑–∞–∫–∞–∑–∞! –ü—Ä–æ–≤–µ—Ä—å—Ç–µ –≤–≤–æ–¥.")
    display_interface()

def on_update_status_clicked(b):
    clear_output(wait=True)
    cafe.update_order_status(order_id_status.value, new_status.value)
    display_interface()

def on_show_menu_clicked(b):
    clear_output(wait=True)
    cafe.show_menu()
    display_interface()

def on_show_orders_clicked(b):
    clear_output(wait=True)
    cafe.show_orders()
    display_interface()

def on_popular_clicked(b):
    clear_output(wait=True)
    cafe.get_popular_dishes()
    display_interface()

def on_revenue_clicked(b):
    clear_output(wait=True)
    cafe.revenue_by_category()
    display_interface()

def on_visualize_clicked(b):
    clear_output(wait=True)
    cafe.visualize_popular_dishes()
    cafe.visualize_revenue_by_category()
    display_interface()

# –ü—Ä–∏–≤—è–∑—ã–≤–∞–µ–º —Ñ—É–Ω–∫—Ü–∏–∏ –∫ –∫–Ω–æ–ø–∫–∞–º
add_dish_btn.on_click(on_add_dish_clicked)
create_order_btn.on_click(on_create_order_clicked)
update_status_btn.on_click(on_update_status_clicked)
show_menu_btn.on_click(on_show_menu_clicked)
show_orders_btn.on_click(on_show_orders_clicked)
popular_btn.on_click(on_popular_clicked)
revenue_btn.on_click(on_revenue_clicked)
visualize_btn.on_click(on_visualize_clicked)

def display_interface():
    """–û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –∏–Ω—Ç–µ—Ä—Ñ–µ–π—Å–∞"""
    print("\n" + "="*50)
    print("üè™ –°–ò–°–¢–ï–ú–ê –£–ü–†–ê–í–õ–ï–ù–ò–Ø –ó–ê–ö–ê–ó–ê–ú–ò –ö–ê–§–ï")
    print("="*50 + "\n")

    # –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –±–ª—é–¥–∞
    print("--- –î–û–ë–ê–í–õ–ï–ù–ò–ï –ë–õ–Æ–î–ê ---")
    display(dish_name, dish_price, dish_category, dish_time, add_dish_btn)

    print("\n--- –°–û–ó–î–ê–ù–ò–ï –ó–ê–ö–ê–ó–ê ---")
    display(order_table, order_dishes, create_order_btn)

    print("\n--- –û–ë–ù–û–í–õ–ï–ù–ò–ï –°–¢–ê–¢–£–°–ê ---")
    display(order_id_status, new_status, update_status_btn)

    print("\n--- –ü–†–û–°–ú–û–¢–† –î–ê–ù–ù–´–• ---")
    display(widgets.HBox([show_menu_btn, show_orders_btn, popular_btn, revenue_btn, visualize_btn]))

# –ó–∞–ø—É—Å–∫–∞–µ–º –∏–Ω—Ç–µ—Ä—Ñ–µ–π—Å
display_interface()

# –î–æ–±–∞–≤–∏–º –ø—Ä–∏–º–µ—Ä –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏–∏
print("\n" + "="*50)
print("üìå –î–û–ë–ê–í–õ–ï–ù–ò–ï –ü–†–ò–ú–ï–†–û–í –î–õ–Ø –¢–ï–°–¢–ò–†–û–í–ê–ù–ò–Ø")
print("="*50)

# –î–æ–±–∞–≤–ª—è–µ–º —Ç–µ—Å—Ç–æ–≤—ã–µ –±–ª—é–¥–∞
cafe.add_dish("–ö–æ—Ñ–µ", 150, "–ù–∞–ø–∏—Ç–∫–∏", 5)
cafe.add_dish("–ß–∞–π", 100, "–ù–∞–ø–∏—Ç–∫–∏", 3)
cafe.add_dish("–ë–æ—Ä—â", 300, "–ì–æ—Ä—è—á–µ–µ", 15)
cafe.add_dish("–ö–æ—Ç–ª–µ—Ç–∞ —Å –ø—é—Ä–µ", 350, "–ì–æ—Ä—è—á–µ–µ", 20)
cafe.add_dish("–¶–µ–∑–∞—Ä—å", 280, "–°–∞–ª–∞—Ç—ã", 10)
cafe.add_dish("–¢–∏—Ä–∞–º–∏—Å—É", 200, "–î–µ—Å–µ—Ä—Ç—ã", 7)
cafe.add_dish("–ö–∞—Ä—Ç–æ—à–∫–∞ —Ñ—Ä–∏", 180, "–ó–∞–∫—É—Å–∫–∏", 8)

# –°–æ–∑–¥–∞–µ–º —Ç–µ—Å—Ç–æ–≤—ã–µ –∑–∞–∫–∞–∑—ã
cafe.create_order(5, [1, 2])  # –°—Ç–æ–ª 5: –ö–æ—Ñ–µ, –ß–∞–π
cafe.create_order(3, [3, 4, 6])  # –°—Ç–æ–ª 3: –ë–æ—Ä—â, –ö–æ—Ç–ª–µ—Ç–∞, –¢–∏—Ä–∞–º–∏—Å—É
cafe.create_order(2, [5, 7, 1])  # –°—Ç–æ–ª 2: –¶–µ–∑–∞—Ä—å, –ö–∞—Ä—Ç–æ—à–∫–∞ —Ñ—Ä–∏, –ö–æ—Ñ–µ

# –û–±–Ω–æ–≤–ª—è–µ–º —Å—Ç–∞—Ç—É—Å—ã –¥–ª—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –≤—ã—Ä—É—á–∫–∏
cafe.update_order_status(1, "–æ–ø–ª–∞—á–µ–Ω")
cafe.update_order_status(2, "–æ–ø–ª–∞—á–µ–Ω")

print("\n‚úÖ –¢–µ—Å—Ç–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ –¥–æ–±–∞–≤–ª–µ–Ω—ã!")
print("üëâ –¢–µ–ø–µ—Ä—å –≤—ã –º–æ–∂–µ—Ç–µ —Ä–∞–±–æ—Ç–∞—Ç—å —Å —Å–∏—Å—Ç–µ–º–æ–π!")
```

```
# @title –°–∏—Å—Ç–µ–º–∞ —É—á–µ—Ç–∞ —É—Å–ø–µ–≤–∞–µ–º–æ—Å—Ç–∏ —Å—Ç—É–¥–µ–Ω—Ç–æ–≤
import pandas as pd
from datetime import datetime
import ipywidgets as widgets
from IPython.display import display, clear_output
import matplotlib.pyplot as plt
import numpy as np

class StudentGradeSystem:
    def __init__(self):
        self.students = {}  # {id: {"name": "", "group": "", "contacts": "", "grades": []}}
        self.grades = []  # —Å–ø–∏—Å–æ–∫ –≤—Å–µ—Ö –æ—Ü–µ–Ω–æ–∫ –¥–ª—è —É–¥–æ–±–Ω–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞
        self.next_student_id = 1

    def add_student(self, name, group, contacts):
        """1. –î–æ–±–∞–≤–ª–µ–Ω–∏–µ —Å—Ç—É–¥–µ–Ω—Ç–∞: ID, –§–ò–û, –≥—Ä—É–ø–ø–∞, –∫–æ–Ω—Ç–∞–∫—Ç—ã"""
        student_id = self.next_student_id
        self.students[student_id] = {
            "name": name,
            "group": group,
            "contacts": contacts,
            "grades": []  # —Å–ø–∏—Å–æ–∫ —Å–ª–æ–≤–∞—Ä–µ–π —Å –æ—Ü–µ–Ω–∫–∞–º–∏
        }
        self.next_student_id += 1
        print(f"‚úÖ –°—Ç—É–¥–µ–Ω—Ç {name} (ID: {student_id}) –¥–æ–±–∞–≤–ª–µ–Ω –≤ —Å–∏—Å—Ç–µ–º—É!")
        return student_id

    def add_grade(self, student_id, subject, grade, date=None):
        """2. –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –æ—Ü–µ–Ω–∫–∏: —Å—Ç—É–¥–µ–Ω—Ç, –ø—Ä–µ–¥–º–µ—Ç, –æ—Ü–µ–Ω–∫–∞, –¥–∞—Ç–∞"""
        if student_id not in self.students:
            print(f"‚ùå –°—Ç—É–¥–µ–Ω—Ç —Å ID {student_id} –Ω–µ –Ω–∞–π–¥–µ–Ω!")
            return False

        if date is None:
            date = datetime.now().strftime("%Y-%m-%d")

        # –ü—Ä–æ–≤–µ—Ä–∫–∞ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ—Å—Ç–∏ –æ—Ü–µ–Ω–∫–∏
        if not (2 <= grade <= 5):
            print(f"‚ùå –û—Ü–µ–Ω–∫–∞ –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –æ—Ç 2 –¥–æ 5!")
            return False

        grade_record = {
            "student_id": student_id,
            "subject": subject,
            "grade": grade,
            "date": date
        }

        self.students[student_id]["grades"].append(grade_record)
        self.grades.append(grade_record)

        print(f"‚úÖ –û—Ü–µ–Ω–∫–∞ {grade} –ø–æ –ø—Ä–µ–¥–º–µ—Ç—É '{subject}' –¥–æ–±–∞–≤–ª–µ–Ω–∞ —Å—Ç—É–¥–µ–Ω—Ç—É {self.students[student_id]['name']}")
        return True

    def calculate_average(self, student_id):
        """3. –†–∞—Å—á–µ—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ –±–∞–ª–ª–∞ –ø–æ —Å—Ç—É–¥–µ–Ω—Ç—É"""
        if student_id not in self.students:
            print(f"‚ùå –°—Ç—É–¥–µ–Ω—Ç —Å ID {student_id} –Ω–µ –Ω–∞–π–¥–µ–Ω!")
            return 0

        grades = self.students[student_id]["grades"]
        if not grades:
            print(f"‚ÑπÔ∏è –£ —Å—Ç—É–¥–µ–Ω—Ç–∞ {self.students[student_id]['name']} –ø–æ–∫–∞ –Ω–µ—Ç –æ—Ü–µ–Ω–æ–∫")
            return 0

        avg = sum(g["grade"] for g in grades) / len(grades)
        print(f"üìä –°—Ä–µ–¥–Ω–∏–π –±–∞–ª–ª —Å—Ç—É–¥–µ–Ω—Ç–∞ {self.students[student_id]['name']}: {avg:.2f}")
        return avg

    def find_students_below_threshold(self, threshold=3.0):
        """4. –ü–æ–∏—Å–∫ —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ —Å –æ—Ü–µ–Ω–∫–æ–π –Ω–∏–∂–µ –∑–∞–¥–∞–Ω–Ω–æ–π (–ø–æ —Å—Ä–µ–¥–Ω–µ–º—É –±–∞–ª–ª—É)"""
        below_threshold = []

        for student_id, info in self.students.items():
            if info["grades"]:
                avg = sum(g["grade"] for g in info["grades"]) / len(info["grades"])
                if avg < threshold:
                    below_threshold.append({
                        "id": student_id,
                        "name": info["name"],
                        "group": info["group"],
                        "avg_grade": avg
                    })

        if below_threshold:
            print(f"\nüìâ –°–¢–£–î–ï–ù–¢–´ –°–û –°–†–ï–î–ù–ò–ú –ë–ê–õ–õ–û–ú –ù–ò–ñ–ï {threshold}:")
            for student in sorted(below_threshold, key=lambda x: x["avg_grade"]):
                print(f"   {student['name']} (–≥—Ä. {student['group']}): {student['avg_grade']:.2f}")
        else:
            print(f"‚úÖ –ù–µ—Ç —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ —Å–æ —Å—Ä–µ–¥–Ω–∏–º –±–∞–ª–ª–æ–º –Ω–∏–∂–µ {threshold}")

        return below_threshold

    def get_students_rating(self):
        """5. –†–µ–π—Ç–∏–Ω–≥ —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ –ø–æ —Å—Ä–µ–¥–Ω–µ–º—É –±–∞–ª–ª—É"""
        rating = []

        for student_id, info in self.students.items():
            if info["grades"]:
                avg = sum(g["grade"] for g in info["grades"]) / len(info["grades"])
                rating.append({
                    "id": student_id,
                    "name": info["name"],
                    "group": info["group"],
                    "avg_grade": avg,
                    "grades_count": len(info["grades"])
                })

        # –°–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ —É–±—ã–≤–∞–Ω–∏—é —Å—Ä–µ–¥–Ω–µ–≥–æ –±–∞–ª–ª–∞
        rating.sort(key=lambda x: x["avg_grade"], reverse=True)

        if rating:
            print("\nüèÜ –†–ï–ô–¢–ò–ù–ì –°–¢–£–î–ï–ù–¢–û–í:")
            print("=" * 60)
            for i, student in enumerate(rating, 1):
                medal = "ü•á" if i == 1 else "ü•à" if i == 2 else "ü•â" if i == 3 else "   "
                print(f"{medal} {i:2}. {student['name']:20} (–≥—Ä. {student['group']:10}) - {student['avg_grade']:.2f}  (–æ—Ü–µ–Ω–æ–∫: {student['grades_count']})")
        else:
            print("üìä –ù–µ—Ç –¥–∞–Ω–Ω—ã—Ö –¥–ª—è —Ä–µ–π—Ç–∏–Ω–≥–∞")

        return rating

    def get_student_info(self, student_id):
        """–ü–æ–ª—É—á–∏—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ —Å—Ç—É–¥–µ–Ω—Ç–µ"""
        if student_id in self.students:
            info = self.students[student_id]
            print(f"\nüë®‚Äçüéì –ò–ù–§–û–†–ú–ê–¶–ò–Ø –û –°–¢–£–î–ï–ù–¢–ï:")
            print(f"   ID: {student_id}")
            print(f"   –§–ò–û: {info['name']}")
            print(f"   –ì—Ä—É–ø–ø–∞: {info['group']}")
            print(f"   –ö–æ–Ω—Ç–∞–∫—Ç—ã: {info['contacts']}")

            if info["grades"]:
                print(f"\n   üìö –û–¶–ï–ù–ö–ò:")
                df = pd.DataFrame(info["grades"])
                print(df.to_string(index=False))

                avg = sum(g["grade"] for g in info["grades"]) / len(info["grades"])
                print(f"\n   –°—Ä–µ–¥–Ω–∏–π –±–∞–ª–ª: {avg:.2f}")
            else:
                print("\n   üìö –û—Ü–µ–Ω–æ–∫ –ø–æ–∫–∞ –Ω–µ—Ç")

            return info
        else:
            print(f"‚ùå –°—Ç—É–¥–µ–Ω—Ç —Å ID {student_id} –Ω–µ –Ω–∞–π–¥–µ–Ω!")
            return None

    def search_by_name(self, name_part):
        """–ü–æ–∏—Å–∫ —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ –ø–æ —á–∞—Å—Ç–∏ –∏–º–µ–Ω–∏"""
        found = []
        name_part = name_part.lower()

        for student_id, info in self.students.items():
            if name_part in info["name"].lower():
                found.append((student_id, info))

        if found:
            print(f"\nüîç –ù–ê–ô–î–ï–ù–ù–´–ï –°–¢–£–î–ï–ù–¢–´ (–ø–æ –∑–∞–ø—Ä–æ—Å—É '{name_part}'):")
            for student_id, info in found:
                print(f"   ID {student_id}: {info['name']} (–≥—Ä. {info['group']})")
        else:
            print(f"üîç –°—Ç—É–¥–µ–Ω—Ç—ã –ø–æ –∑–∞–ø—Ä–æ—Å—É '{name_part}' –Ω–µ –Ω–∞–π–¥–µ–Ω—ã")

        return found

    def visualize_grade_distribution(self):
        """–í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –æ—Ü–µ–Ω–æ–∫"""
        if not self.grades:
            print("üìä –ù–µ—Ç –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏")
            return

        grades_df = pd.DataFrame(self.grades)
        grade_counts = grades_df['grade'].value_counts().sort_index()

        plt.figure(figsize=(10, 6))
        bars = plt.bar(grade_counts.index, grade_counts.values,
                       color=['red', 'orange', 'yellow', 'green'])
        plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –æ—Ü–µ–Ω–æ–∫', fontsize=16)
        plt.xlabel('–û—Ü–µ–Ω–∫–∞', fontsize=12)
        plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ', fontsize=12)

        # –î–æ–±–∞–≤–ª—è–µ–º –ø–æ–¥–ø–∏—Å–∏ –Ω–∞ —Å—Ç–æ–ª–±—Ü—ã
        for bar in bars:
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2., height,
                    f'{int(height)}', ha='center', va='bottom')

        plt.grid(axis='y', alpha=0.3)
        plt.show()

    def visualize_group_performance(self):
        """–í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —É—Å–ø–µ–≤–∞–µ–º–æ—Å—Ç–∏ –ø–æ –≥—Ä—É–ø–ø–∞–º"""
        group_stats = {}

        for student_id, info in self.students.items():
            if info["grades"]:
                group = info["group"]
                avg = sum(g["grade"] for g in info["grades"]) / len(info["grades"])

                if group not in group_stats:
                    group_stats[group] = {"sum": 0, "count": 0, "students": []}

                group_stats[group]["sum"] += avg
                group_stats[group]["count"] += 1
                group_stats[group]["students"].append(avg)

        if not group_stats:
            print("üìä –ù–µ—Ç –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏")
            return

        groups = list(group_stats.keys())
        averages = [stats["sum"] / stats["count"] for stats in group_stats.values()]

        plt.figure(figsize=(12, 6))

        # –°—Ç–æ–ª–±—á–∞—Ç–∞—è –¥–∏–∞–≥—Ä–∞–º–º–∞
        plt.subplot(1, 2, 1)
        bars = plt.bar(groups, averages, color='skyblue')
        plt.title('–°—Ä–µ–¥–Ω–∏–π –±–∞–ª–ª –ø–æ –≥—Ä—É–ø–ø–∞–º', fontsize=14)
        plt.xlabel('–ì—Ä—É–ø–ø–∞')
        plt.ylabel('–°—Ä–µ–¥–Ω–∏–π –±–∞–ª–ª')
        plt.ylim(2, 5)

        for bar, avg in zip(bars, averages):
            plt.text(bar.get_x() + bar.get_width()/2., bar.get_height(),
                    f'{avg:.2f}', ha='center', va='bottom')

        # –Ø—â–∏–∫ —Å —É—Å–∞–º–∏ (box plot) –¥–ª—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
        plt.subplot(1, 2, 2)
        data = [group_stats[group]["students"] for group in groups]
        plt.boxplot(data, labels=groups)
        plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —É—Å–ø–µ–≤–∞–µ–º–æ—Å—Ç–∏ –ø–æ –≥—Ä—É–ø–ø–∞–º', fontsize=14)
        plt.xlabel('–ì—Ä—É–ø–ø–∞')
        plt.ylabel('–°—Ä–µ–¥–Ω–∏–π –±–∞–ª–ª —Å—Ç—É–¥–µ–Ω—Ç–æ–≤')
        plt.grid(alpha=0.3)

        plt.tight_layout()
        plt.show()

    def export_to_dataframe(self):
        """–≠–∫—Å–ø–æ—Ä—Ç –¥–∞–Ω–Ω—ã—Ö –≤ pandas DataFrame"""
        students_list = []
        for student_id, info in self.students.items():
            avg = 0
            if info["grades"]:
                avg = sum(g["grade"] for g in info["grades"]) / len(info["grades"])

            students_list.append({
                "ID": student_id,
                "–§–ò–û": info["name"],
                "–ì—Ä—É–ø–ø–∞": info["group"],
                "–ö–æ–Ω—Ç–∞–∫—Ç—ã": info["contacts"],
                "–ö–æ–ª-–≤–æ –æ—Ü–µ–Ω–æ–∫": len(info["grades"]),
                "–°—Ä–µ–¥–Ω–∏–π –±–∞–ª–ª": round(avg, 2)
            })

        return pd.DataFrame(students_list)

# @title –ò–Ω—Ç–µ—Ä—Ñ–µ–π—Å –¥–ª—è —Ä–∞–±–æ—Ç—ã —Å —Å–∏—Å—Ç–µ–º–æ–π
# –°–æ–∑–¥–∞–µ–º —ç–∫–∑–µ–º–ø–ª—è—Ä —Å–∏—Å—Ç–µ–º—ã
system = StudentGradeSystem()

print("=" * 60)
print("üè´ –°–ò–°–¢–ï–ú–ê –£–ß–ï–¢–ê –£–°–ü–ï–í–ê–ï–ú–û–°–¢–ò –°–¢–£–î–ï–ù–¢–û–í")
print("=" * 60)

# –°–æ–∑–¥–∞–µ–º –≤–∏–¥–∂–µ—Ç—ã
# –î–æ–±–∞–≤–ª–µ–Ω–∏–µ —Å—Ç—É–¥–µ–Ω—Ç–∞
add_student_btn = widgets.Button(description="‚ûï –î–æ–±–∞–≤–∏—Ç—å —Å—Ç—É–¥–µ–Ω—Ç–∞", button_style='success')
student_name = widgets.Text(placeholder="–§–ò–û —Å—Ç—É–¥–µ–Ω—Ç–∞")
student_group = widgets.Text(placeholder="–ì—Ä—É–ø–ø–∞")
student_contacts = widgets.Text(placeholder="–ö–æ–Ω—Ç–∞–∫—Ç—ã (—Ç–µ–ª–µ—Ñ–æ–Ω/email)")

# –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –æ—Ü–µ–Ω–∫–∏
add_grade_btn = widgets.Button(description="üìù –î–æ–±–∞–≤–∏—Ç—å –æ—Ü–µ–Ω–∫—É", button_style='info')
grade_student_id = widgets.IntText(placeholder="ID —Å—Ç—É–¥–µ–Ω—Ç–∞")
grade_subject = widgets.Text(placeholder="–ü—Ä–µ–¥–º–µ—Ç")
grade_value = widgets.Dropdown(options=[2,3,4,5], value=4, description='–û—Ü–µ–Ω–∫–∞:')
grade_date = widgets.DatePicker(description='–î–∞—Ç–∞:')

# –ê–Ω–∞–ª–∏—Ç–∏–∫–∞
calc_avg_btn = widgets.Button(description="üìä –°—Ä–µ–¥–Ω–∏–π –±–∞–ª–ª")
avg_student_id = widgets.IntText(placeholder="ID —Å—Ç—É–¥–µ–Ω—Ç–∞")

find_below_btn = widgets.Button(description="üìâ –ü–æ–∏—Å–∫ –Ω–∏–∂–µ –ø–æ—Ä–æ–≥–∞")
below_threshold = widgets.FloatText(value=3.0, description='–ü–æ—Ä–æ–≥:')

rating_btn = widgets.Button(description="üèÜ –†–µ–π—Ç–∏–Ω–≥ —Å—Ç—É–¥–µ–Ω—Ç–æ–≤")

# –ü–æ–∏—Å–∫ –∏ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è
search_btn = widgets.Button(description="üîç –ü–æ–∏—Å–∫ –ø–æ –∏–º–µ–Ω–∏")
search_name = widgets.Text(placeholder="–ß–∞—Å—Ç—å –∏–º–µ–Ω–∏")

info_btn = widgets.Button(description="üë®‚Äçüéì –ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ —Å—Ç—É–¥–µ–Ω—Ç–µ")
info_id = widgets.IntText(placeholder="ID —Å—Ç—É–¥–µ–Ω—Ç–∞")

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è
visualize_btn = widgets.Button(description="üìà –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è", button_style='warning')
export_btn = widgets.Button(description="üìã –≠–∫—Å–ø–æ—Ä—Ç –≤ DataFrame")

def on_add_student_clicked(b):
    clear_output(wait=True)
    if student_name.value and student_group.value:
        system.add_student(student_name.value, student_group.value, student_contacts.value)
    else:
        print("‚ùå –ó–∞–ø–æ–ª–Ω–∏—Ç–µ –§–ò–û –∏ –≥—Ä—É–ø–ø—É!")
    display_interface()

def on_add_grade_clicked(b):
    clear_output(wait=True)
    if grade_student_id.value and grade_subject.value:
        date_str = grade_date.value.strftime("%Y-%m-%d") if grade_date.value else None
        system.add_grade(grade_student_id.value, grade_subject.value, grade_value.value, date_str)
    else:
        print("‚ùå –ó–∞–ø–æ–ª–Ω–∏—Ç–µ ID —Å—Ç—É–¥–µ–Ω—Ç–∞ –∏ –ø—Ä–µ–¥–º–µ—Ç!")
    display_interface()

def on_calc_avg_clicked(b):
    clear_output(wait=True)
    if avg_student_id.value:
        system.calculate_average(avg_student_id.value)
    else:
        print("‚ùå –í–≤–µ–¥–∏—Ç–µ ID —Å—Ç—É–¥–µ–Ω—Ç–∞!")
    display_interface()

def on_find_below_clicked(b):
    clear_output(wait=True)
    system.find_students_below_threshold(below_threshold.value)
    display_interface()

def on_rating_clicked(b):
    clear_output(wait=True)
    system.get_students_rating()
    display_interface()

def on_search_clicked(b):
    clear_output(wait=True)
    if search_name.value:
        system.search_by_name(search_name.value)
    else:
        print("‚ùå –í–≤–µ–¥–∏—Ç–µ –∏–º—è –¥–ª—è –ø–æ–∏—Å–∫–∞!")
    display_interface()

def on_info_clicked(b):
    clear_output(wait=True)
    if info_id.value:
        system.get_student_info(info_id.value)
    else:
        print("‚ùå –í–≤–µ–¥–∏—Ç–µ ID —Å—Ç—É–¥–µ–Ω—Ç–∞!")
    display_interface()

def on_visualize_clicked(b):
    clear_output(wait=True)
    system.visualize_grade_distribution()
    system.visualize_group_performance()
    display_interface()

def on_export_clicked(b):
    clear_output(wait=True)
    df = system.export_to_dataframe()
    print("\nüìã –î–ê–ù–ù–´–ï –°–¢–£–î–ï–ù–¢–û–í:")
    display(df)
    display_interface()

# –ü—Ä–∏–≤—è–∑—ã–≤–∞–µ–º —Ñ—É–Ω–∫—Ü–∏–∏
add_student_btn.on_click(on_add_student_clicked)
add_grade_btn.on_click(on_add_grade_clicked)
calc_avg_btn.on_click(on_calc_avg_clicked)
find_below_btn.on_click(on_find_below_clicked)
rating_btn.on_click(on_rating_clicked)
search_btn.on_click(on_search_clicked)
info_btn.on_click(on_info_clicked)
visualize_btn.on_click(on_visualize_clicked)
export_btn.on_click(on_export_clicked)

def display_interface():
    """–û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –∏–Ω—Ç–µ—Ä—Ñ–µ–π—Å–∞"""
    print("\n" + "="*60)
    print("üè´ –°–ò–°–¢–ï–ú–ê –£–ß–ï–¢–ê –£–°–ü–ï–í–ê–ï–ú–û–°–¢–ò –°–¢–£–î–ï–ù–¢–û–í")
    print("="*60 + "\n")

    print("--- –î–û–ë–ê–í–õ–ï–ù–ò–ï –°–¢–£–î–ï–ù–¢–ê ---")
    display(student_name, student_group, student_contacts, add_student_btn)

    print("\n--- –î–û–ë–ê–í–õ–ï–ù–ò–ï –û–¶–ï–ù–ö–ò ---")
    display(grade_student_id, grade_subject, grade_value, grade_date, add_grade_btn)

    print("\n--- –ê–ù–ê–õ–ò–¢–ò–ö–ê ---")
    display(widgets.HBox([avg_student_id, calc_avg_btn]))
    display(widgets.HBox([below_threshold, find_below_btn]))
    display(rating_btn)

    print("\n--- –ü–û–ò–°–ö –ò –ò–ù–§–û–†–ú–ê–¶–ò–Ø ---")
    display(widgets.HBox([search_name, search_btn]))
    display(widgets.HBox([info_id, info_btn]))

    print("\n--- –í–ò–ó–£–ê–õ–ò–ó–ê–¶–ò–Ø –ò –≠–ö–°–ü–û–†–¢ ---")
    display(widgets.HBox([visualize_btn, export_btn]))

# –ó–∞–ø—É—Å–∫–∞–µ–º –∏–Ω—Ç–µ—Ä—Ñ–µ–π—Å
display_interface()

# –î–æ–±–∞–≤–ª—è–µ–º —Ç–µ—Å—Ç–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ
print("\n" + "="*60)
print("üìå –î–û–ë–ê–í–õ–ï–ù–ò–ï –¢–ï–°–¢–û–í–´–• –î–ê–ù–ù–´–•")
print("="*60)

# –î–æ–±–∞–≤–ª—è–µ–º —Å—Ç—É–¥–µ–Ω—Ç–æ–≤
system.add_student("–ò–≤–∞–Ω–æ–≤ –ò–≤–∞–Ω –ò–≤–∞–Ω–æ–≤–∏—á", "–ò–°-21", "ivanov@mail.com")
system.add_student("–ü–µ—Ç—Ä–æ–≤–∞ –ê–Ω–Ω–∞ –°–µ—Ä–≥–µ–µ–≤–Ω–∞", "–ò–°-21", "+7-999-123-45-67")
system.add_student("–°–∏–¥–æ—Ä–æ–≤ –ü–µ—Ç—Ä –ê–ª–µ–∫—Å–µ–µ–≤–∏—á", "–ë–ò-22", "sidorov@yandex.ru")
system.add_student("–ö–æ–∑–ª–æ–≤–∞ –ú–∞—Ä–∏—è –î–º–∏—Ç—Ä–∏–µ–≤–Ω–∞", "–ò–°-21", "+7-911-987-65-43")
system.add_student("–°–º–∏—Ä–Ω–æ–≤ –ê–ª–µ–∫—Å–µ–π –í–ª–∞–¥–∏–º–∏—Ä–æ–≤–∏—á", "–ë–ò-22", "smirnov@gmail.com")

# –î–æ–±–∞–≤–ª—è–µ–º –æ—Ü–µ–Ω–∫–∏
system.add_grade(1, "–ú–∞—Ç–µ–º–∞—Ç–∏–∫–∞", 5, "2024-02-01")
system.add_grade(1, "–§–∏–∑–∏–∫–∞", 4, "2024-02-05")
system.add_grade(1, "–ü—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞–Ω–∏–µ", 5, "2024-02-10")
system.add_grade(1, "–ú–∞—Ç–µ–º–∞—Ç–∏–∫–∞", 5, "2024-02-15")

system.add_grade(2, "–ú–∞—Ç–µ–º–∞—Ç–∏–∫–∞", 4, "2024-02-01")
system.add_grade(2, "–§–∏–∑–∏–∫–∞", 5, "2024-02-05")
system.add_grade(2, "–ü—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞–Ω–∏–µ", 5, "2024-02-10")
system.add_grade(2, "–ê–Ω–≥–ª–∏–π—Å–∫–∏–π", 5, "2024-02-12")

system.add_grade(3, "–ú–∞—Ç–µ–º–∞—Ç–∏–∫–∞", 3, "2024-02-01")
system.add_grade(3, "–§–∏–∑–∏–∫–∞", 3, "2024-02-05")
system.add_grade(3, "–•–∏–º–∏—è", 4, "2024-02-08")

system.add_grade(4, "–ú–∞—Ç–µ–º–∞—Ç–∏–∫–∞", 4, "2024-02-01")
system.add_grade(4, "–ü—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞–Ω–∏–µ", 4, "2024-02-10")
system.add_grade(4, "–ê–Ω–≥–ª–∏–π—Å–∫–∏–π", 5, "2024-02-12")

system.add_grade(5, "–ú–∞—Ç–µ–º–∞—Ç–∏–∫–∞", 3, "2024-02-01")
system.add_grade(5, "–§–∏–∑–∏–∫–∞", 3, "2024-02-05")
system.add_grade(5, "–•–∏–º–∏—è", 3, "2024-02-08")

print("\n‚úÖ –¢–µ—Å—Ç–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ –¥–æ–±–∞–≤–ª–µ–Ω—ã!")
print("üëâ –¢–µ–ø–µ—Ä—å –≤—ã –º–æ–∂–µ—Ç–µ —Ä–∞–±–æ—Ç–∞—Ç—å —Å —Å–∏—Å—Ç–µ–º–æ–π!")
```
