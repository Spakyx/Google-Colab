```
import random

#–Ω–∞—Å—Ç—Ä–æ–π–∫–∞ –¥–∏–∞–ø–∞–∑–æ–Ω–∞ –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –∫–æ–ª-—Ç–≤–∞ –ø–æ–ø—ã—Ç–æ–∫
max_attempts = 7
min_range = 1
max_range = 100

secret_number = random.randint(min_range, max_range)
attempts = 1

print(f"–Ø –∑–∞–≥–∞–¥–∞–ª —á–∏—Å–ª–æ –æ—Ç {min_range} –¥–æ {max_range}")
print(f"–£ –≤–∞—Å –µ—Å—Ç—å {max_attempts} –ø–æ–ø—ã—Ç–æ–∫!")

while True:

  guess = int(input("–í–∞—à–µ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏–µ: "))

  if guess == secret_number:
     print(f"\nüéâ –ü–æ–∑–¥—Ä–∞–≤–ª—è—é! –í—ã —É–≥–∞–¥–∞–ª–∏ —á–∏—Å–ª–æ {secret_number}!")
     print(f"üìä –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–æ –ø–æ–ø—ã—Ç–æ–∫: {attempts}")
     break
  elif guess < secret_number:
      print("üìà –ó–∞–≥–∞–¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ –ë–û–õ–¨–®–ï")
  else:
      print("üìâ –ó–∞–≥–∞–¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ –ú–ï–ù–¨–®–ï")


  if attempts < max_attempts:
    print(f"üí° –û—Å—Ç–∞–ª–æ—Å—å –ø–æ–ø—ã—Ç–æ–∫: {max_attempts - attempts}")
    attempts +=1
  else:
    print(f"\nüíÄ Game Over! –í—ã –∏—Å—á–µ—Ä–ø–∞–ª–∏ –≤—Å–µ –ø–æ–ø—ã—Ç–∫–∏.")
    print(f"üî¢ –ó–∞–≥–∞–¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ –±—ã–ª–æ: {secret_number}")
    break
```

```
import random
import string
import re

def generate_password(length, use_letters=True, use_digits=True, use_specials=True, count=1):
    """
    –ì–µ–Ω–µ—Ä–∏—Ä—É–µ—Ç –æ–¥–∏–Ω –∏–ª–∏ –Ω–µ—Å–∫–æ–ª—å–∫–æ –ø–∞—Ä–æ–ª–µ–π —Å –∑–∞–¥–∞–Ω–Ω—ã–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏
    """
    characters = ""

    if use_letters:
        characters += string.ascii_letters  # –ó–∞–≥–ª–∞–≤–Ω—ã–µ –∏ —Å—Ç—Ä–æ—á–Ω—ã–µ –±—É–∫–≤—ã

    if use_digits:
        characters += string.digits  # –¶–∏—Ñ—Ä—ã

    if use_specials:
        characters += "!@#$%^&*()_+-=[]{}|;:,.<>?"  # –°–ø–µ—Ü–∏–∞–ª—å–Ω—ã–µ —Å–∏–º–≤–æ–ª—ã


    if not characters:
        return ["–û—à–∏–±–∫–∞: –í—ã–±–µ—Ä–∏—Ç–µ —Ö–æ—Ç—è –±—ã –æ–¥–∏–Ω —Ç–∏–ø —Å–∏–º–≤–æ–ª–æ–≤!"]


    passwords = []
    for _ in range(count):
        password = ''.join(random.choice(characters) for _ in range(length))
        passwords.append(password)

    return passwords

def check_password_strength(password):
    """
    –ü—Ä–æ–≤–µ—Ä—è–µ—Ç —Å–ª–æ–∂–Ω–æ—Å—Ç—å –ø–∞—Ä–æ–ª—è –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –æ—Ü–µ–Ω–∫—É –∏ —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏
    """
    score = 0
    feedback = []


    if len(password) >= 12:
        score += 2
        feedback.append("–î–ª–∏–Ω–∞ –æ—Ç–ª–∏—á–Ω–∞—è (12+ —Å–∏–º–≤–æ–ª–æ–≤)")
    elif len(password) >= 8:
        score += 1
        feedback.append("–î–ª–∏–Ω–∞ —Ö–æ—Ä–æ—à–∞—è (8-11 —Å–∏–º–≤–æ–ª–æ–≤)")
    else:
        feedback.append("–î–ª–∏–Ω–∞ —Å–ª–∏—à–∫–æ–º –∫–æ—Ä–æ—Ç–∫–∞—è (–º–µ–Ω–µ–µ 8 —Å–∏–º–≤–æ–ª–æ–≤)")


    if re.search(r'[A-Z]', password):
        score += 1
        feedback.append("–ï—Å—Ç—å –∑–∞–≥–ª–∞–≤–Ω—ã–µ –±—É–∫–≤—ã")
    else:
        feedback.append("–ù–µ—Ç –∑–∞–≥–ª–∞–≤–Ω—ã—Ö –±—É–∫–≤")


    if re.search(r'[a-z]', password):
        score += 1
        feedback.append("–ï—Å—Ç—å —Å—Ç—Ä–æ—á–Ω—ã–µ –±—É–∫–≤—ã")
    else:
        feedback.append("–ù–µ—Ç —Å—Ç—Ä–æ—á–Ω—ã—Ö –±—É–∫–≤")


    if re.search(r'\d', password):
        score += 1
        feedback.append("–ï—Å—Ç—å —Ü–∏—Ñ—Ä—ã")
    else:
        feedback.append("–ù–µ—Ç —Ü–∏—Ñ—Ä")


    if re.search(r'[!@#$%^&*()_+\-=\[\]{}|;:,.<>?]', password):
        score += 1
        feedback.append("–ï—Å—Ç—å —Å–ø–µ—Ü–∏–∞–ª—å–Ω—ã–µ —Å–∏–º–≤–æ–ª—ã")
    else:
        feedback.append("–ù–µ—Ç —Å–ø–µ—Ü–∏–∞–ª—å–Ω—ã—Ö —Å–∏–º–≤–æ–ª–æ–≤")


    if score >= 5:
        strength = "–û–ß–ï–ù–¨ –ù–ê–î–ï–ñ–ù–´–ô"
    elif score >= 4:
        strength = "–ù–ê–î–ï–ñ–ù–´–ô"
    elif score >= 3:
        strength = "–°–†–ï–î–ù–ò–ô"
    else:
        strength = "–°–õ–ê–ë–´–ô"

    return strength, score, feedback

def main():
    print("=" * 60)
    print("           –ì–ï–ù–ï–†–ê–¢–û–† –ë–ï–ó–û–ü–ê–°–ù–´–• –ü–ê–†–û–õ–ï–ô")
    print("=" * 60)


    try:
        length = int(input("–í–≤–µ–¥–∏—Ç–µ –¥–ª–∏–Ω—É –ø–∞—Ä–æ–ª—è (—Ä–µ–∫–æ–º–µ–Ω–¥—É–µ—Ç—Å—è 8-16): "))
        if length < 1:
            print("–î–ª–∏–Ω–∞ –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–º —á–∏—Å–ª–æ–º. –£—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–æ –∑–Ω–∞—á–µ–Ω–∏–µ 8.")
            length = 8
    except ValueError:
        print("–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –≤–≤–æ–¥. –£—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–∞ –¥–ª–∏–Ω–∞ 8.")
        length = 8

    print("\n–í—ã–±–µ—Ä–∏—Ç–µ —Ç–∏–ø—ã —Å–∏–º–≤–æ–ª–æ–≤:")
    use_letters = input("–ò—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –±—É–∫–≤—ã? (y/n): ").lower() == 'y'
    use_digits = input("–ò—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ü–∏—Ñ—Ä—ã? (y/n): ").lower() == 'y'
    use_specials = input("–ò—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Å–ø–µ—Ü—Å–∏–º–≤–æ–ª—ã? (y/n): ").lower() == 'y'


    if not (use_letters or use_digits or use_specials):
        print("\n–ù–µ –≤—ã–±—Ä–∞–Ω –Ω–∏ –æ–¥–∏–Ω —Ç–∏–ø —Å–∏–º–≤–æ–ª–æ–≤. –ë—É–¥—É—Ç –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω—ã –±—É–∫–≤—ã.")
        use_letters = True

    try:
        count = int(input("\n–°–∫–æ–ª—å–∫–æ –ø–∞—Ä–æ–ª–µ–π —Å–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞—Ç—å? (1-10): "))
        if count < 1:
            count = 1
        elif count > 10:
            count = 10
            print("–°–ª–∏—à–∫–æ–º –º–Ω–æ–≥–æ –ø–∞—Ä–æ–ª–µ–π. –°–≥–µ–Ω–µ—Ä–∏—Ä—É—é 10.")
    except ValueError:
        print("–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –≤–≤–æ–¥. –ë—É–¥–µ—Ç —Å–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞–Ω 1 –ø–∞—Ä–æ–ª—å.")
        count = 1

    print("\n" + "-" * 60)
    print("–ì–ï–ù–ï–†–ê–¶–ò–Ø –ü–ê–†–û–õ–ï–ô...")
    print("-" * 60)


    passwords = generate_password(length, use_letters, use_digits, use_specials, count)


    for i, password in enumerate(passwords, 1):
        print(f"\n–ü–∞—Ä–æ–ª—å {i}: {password}")


        strength, score, feedback = check_password_strength(password)
        print(f"–°–ª–æ–∂–Ω–æ—Å—Ç—å: {strength} (–æ—Ü–µ–Ω–∫–∞: {score}/6)")


        if i == 1:
            print("\n–î–µ—Ç–∞–ª—å–Ω—ã–π –∞–Ω–∞–ª–∏–∑:")
            for item in feedback:
                print(f"  {item}")

    print("\n" + "=" * 60)
    print("–°–û–í–ï–¢–´ –ü–û –ë–ï–ó–û–ü–ê–°–ù–û–°–¢–ò:")
    print("‚Ä¢ –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ä–∞–∑–Ω—ã–µ –ø–∞—Ä–æ–ª–∏ –¥–ª—è —Ä–∞–∑–Ω—ã—Ö —Å–µ—Ä–≤–∏—Å–æ–≤")
    print("‚Ä¢ –ú–µ–Ω—è–π—Ç–µ –ø–∞—Ä–æ–ª–∏ –∫–∞–∂–¥—ã–µ 3-6 –º–µ—Å—è—Ü–µ–≤")
    print("‚Ä¢ –ù–µ —Å–æ–æ–±—â–∞–π—Ç–µ –ø–∞—Ä–æ–ª–∏ –¥—Ä—É–≥–∏–º –ª—é–¥—è–º")
    print("‚Ä¢ –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ –º–µ–Ω–µ–¥–∂–µ—Ä –ø–∞—Ä–æ–ª–µ–π –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è")
    print("=" * 60)


if __name__ == "__main__":
    main()
```
