```
import re

EMAIL_PATTERN = r'[\w.-]+@[\w-]+\.(com|net|org)$'
```

```
def validate_age():
    while True:
        age_input = input("–í–≤–µ–¥–∏—Ç–µ –≤–∞—à –≤–æ–∑—Ä–∞—Å—Ç (—Ü–µ–ª–æ–µ —á–∏—Å–ª–æ –æ—Ç 0 –¥–æ 120): ")
        try:

            age = int(age_input)


            if 0 <= age <= 120:
                print(f"‚úÖ –í–æ–∑—Ä–∞—Å—Ç –ø—Ä–∏–Ω—è—Ç: {age}")
                return age
            else:
                print("‚ùå –í–æ–∑—Ä–∞—Å—Ç –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ –æ—Ç 0 –¥–æ 120.")
        except ValueError:
            print("‚ùå –û—à–∏–±–∫–∞ –≤–≤–æ–¥–∞. –ü–æ–∂–∞–ª—É–π—Å—Ç–∞, –≤–≤–µ–¥–∏—Ç–µ **—Ü–µ–ª–æ–µ —á–∏—Å–ª–æ**.")
```

```
def validate_email():
    while True:
        email = input("–í–≤–µ–¥–∏—Ç–µ –≤–∞—à email (—Ñ–æ—Ä–º–∞—Ç: user@domain.com): ").strip()

        if re.fullmatch(EMAIL_PATTERN, email):
            print(f"‚úÖ Email –ø—Ä–∏–Ω—è—Ç: {email}")
            return email
        else:
            print("‚ùå –ù–µ–≤–µ—Ä–Ω—ã–π —Ñ–æ—Ä–º–∞—Ç email. –£–±–µ–¥–∏—Ç–µ—Å—å, —á—Ç–æ –∞–¥—Ä–µ—Å —Å–æ–¥–µ—Ä–∂–∏—Ç @ –∏ –¥–æ–º–µ–Ω .com/.net/.org")
```

```
print("--- –ü—Ä–æ–≤–µ—Ä–∫–∞ –≤–æ–∑—Ä–∞—Å—Ç–∞ ---")
user_age = validate_age()

print("\n--- –ü—Ä–æ–≤–µ—Ä–∫–∞ email ---")
user_email = validate_email()

print("\n--- –°–≤–æ–¥–∫–∞ ---")
print(f"–í–≤–æ–¥ –∑–∞–≤–µ—Ä—à–µ–Ω. –í–æ–∑—Ä–∞—Å—Ç: {user_age}, Email: {user_email}")
```

```
import re

def validate_inn():
    """
    –§—É–Ω–∫—Ü–∏—è –¥–ª—è –≤–∞–ª–∏–¥–∞—Ü–∏–∏ –ò–ù–ù —Ñ–∏–∑–∏—á–µ—Å–∫–æ–≥–æ –ª–∏—Ü–∞.
    –ò–ù–ù –¥–æ–ª–∂–µ–Ω —Å–æ—Å—Ç–æ—è—Ç—å —Ä–æ–≤–Ω–æ –∏–∑ 12 —Ü–∏—Ñ—Ä.
    """

    print("=" * 50)
    print(" –ü–†–û–í–ï–†–ö–ê –ò–ù–ù (–¥–ª—è —Ñ–∏–∑–∏—á–µ—Å–∫–∏—Ö –ª–∏—Ü)")
    print("=" * 50)
    print("–ò–ù–ù –¥–æ–ª–∂–µ–Ω —Å–æ—Å—Ç–æ—è—Ç—å —Ä–æ–≤–Ω–æ –∏–∑ 12 —Ü–∏—Ñ—Ä")
    print("-" * 50)

    while True:
        try:
            inn = input("–í–≤–µ–¥–∏—Ç–µ –ò–ù–ù (12 —Ü–∏—Ñ—Ä): ").strip()

            if not inn:
                raise ValueError("–ò–ù–ù –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å –ø—É—Å—Ç—ã–º!")

            if not re.match(r'^\d{12}$', inn):
                raise ValueError("–ò–ù–ù –¥–æ–ª–∂–µ–Ω —Å–æ–¥–µ—Ä–∂–∞—Ç—å —Ä–æ–≤–Ω–æ 12 —Ü–∏—Ñ—Ä (–±–µ–∑ –±—É–∫–≤ –∏ –¥—Ä—É–≥–∏—Ö —Å–∏–º–≤–æ–ª–æ–≤)!")
            if ' ' in inn:
                raise ValueError("–ò–ù–ù –Ω–µ –¥–æ–ª–∂–µ–Ω —Å–æ–¥–µ—Ä–∂–∞—Ç—å –ø—Ä–æ–±–µ–ª–æ–≤!")

            print("\n –ò–ù–ù —É—Å–ø–µ—à–Ω–æ –ø—Ä–æ—à–µ–ª –≤–∞–ª–∏–¥–∞—Ü–∏—é!")
            print(f"   –í–≤–µ–¥–µ–Ω–Ω—ã–π –ò–ù–ù: {inn}")

            formatted_inn = f"{inn[:3]} {inn[3:6]} {inn[6:9]} {inn[9:]}"
            print(f"   –§–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π: {formatted_inn}")

            return inn

        except ValueError as e:
            print(f"–û—à–∏–±–∫–∞: {e}")
            print("–ü–æ–ø—Ä–æ–±—É–π—Ç–µ —Å–Ω–æ–≤–∞...\n")
        except Exception as e:
            print(f"–ù–µ–ø—Ä–µ–¥–≤–∏–¥–µ–Ω–Ω–∞—è –æ—à–∏–±–∫–∞: {e}")
            print("–ü–æ–ø—Ä–æ–±—É–π—Ç–µ —Å–Ω–æ–≤–∞...\n")

def validate_inn_with_check_digits():
    """
    –†–∞—Å—à–∏—Ä–µ–Ω–Ω–∞—è –≤–µ—Ä—Å–∏—è —Å –ø—Ä–æ–≤–µ—Ä–∫–æ–π –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω—ã—Ö —Ü–∏—Ñ—Ä –ò–ù–ù.
    –ò–ù–ù —Ñ–∏–∑–∏—á–µ—Å–∫–æ–≥–æ –ª–∏—Ü–∞: 12 —Ü–∏—Ñ—Ä, –ø–æ—Å–ª–µ–¥–Ω–∏–µ 2 - –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω—ã–µ.
    """

    print("=" * 50)
    print("–ü–†–û–í–ï–†–ö–ê –ò–ù–ù (—Å –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω—ã–º–∏ —Ü–∏—Ñ—Ä–∞–º–∏)")
    print("=" * 50)

    def calculate_control_digits(inn):
        """–†–∞—Å—á–µ—Ç –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω—ã—Ö —Ü–∏—Ñ—Ä –¥–ª—è –ò–ù–ù"""
        weights_11 = [7, 2, 4, 10, 3, 5, 9, 4, 6, 8]
        weights_12 = [3, 7, 2, 4, 10, 3, 5, 9, 4, 6, 8]

        control_11 = sum(int(inn[i]) * weights_11[i] for i in range(10)) % 11
        control_11 = 0 if control_11 == 10 else control_11

        control_12 = sum(int(inn[i]) * weights_12[i] for i in range(11)) % 11
        control_12 = 0 if control_12 == 10 else control_12

        return control_11, control_12

    while True:
        try:
            inn = input("–í–≤–µ–¥–∏—Ç–µ –ò–ù–ù –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ (12 —Ü–∏—Ñ—Ä): ").strip()

            if not re.match(r'^\d{12}$', inn):
                raise ValueError("–ò–ù–ù –¥–æ–ª–∂–µ–Ω —Å–æ–¥–µ—Ä–∂–∞—Ç—å —Ä–æ–≤–Ω–æ 12 —Ü–∏—Ñ—Ä!")

            control_11, control_12 = calculate_control_digits(inn)

            if int(inn[10]) != control_11 or int(inn[11]) != control_12:
                raise ValueError(f"–ù–µ–≤–µ—Ä–Ω—ã–µ –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω—ã–µ —Ü–∏—Ñ—Ä—ã! –î–æ–ª–∂–Ω—ã –±—ã—Ç—å: {control_11}{control_12}")

            print("\n–ò–ù–ù –∫–æ—Ä—Ä–µ–∫—Ç–µ–Ω –∏ –ø—Ä–æ—à–µ–ª –ø—Ä–æ–≤–µ—Ä–∫—É –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω—ã—Ö —Ü–∏—Ñ—Ä!")
            print(f"   –ò–ù–ù: {inn[:3]} {inn[3:6]} {inn[6:9]} {inn[9:]}")

            region = inn[:2]
            print(f"   –†–µ–≥–∏–æ–Ω: {region}")
            print(f"   –ö–æ–Ω—Ç—Ä–æ–ª—å–Ω—ã–µ —Ü–∏—Ñ—Ä—ã: {inn[10:12]}")

            return inn

        except ValueError as e:
            print(f"–û—à–∏–±–∫–∞: {e}")
            print("–ü–æ–ø—Ä–æ–±—É–π—Ç–µ —Å–Ω–æ–≤–∞...\n")

def validate_inn_menu():
    """
    –ú–µ–Ω—é –¥–ª—è –≤—ã–±–æ—Ä–∞ –≤–µ—Ä—Å–∏–∏ –ø—Ä–æ–≤–µ—Ä–∫–∏ –ò–ù–ù
    """
    while True:
        print("\n" + "=" * 50)
        print("–í–´–ë–ï–†–ò–¢–ï –†–ï–ñ–ò–ú –ü–†–û–í–ï–†–ö–ò –ò–ù–ù:")
        print("=" * 50)
        print("1. –ë–∞–∑–æ–≤–∞—è –ø—Ä–æ–≤–µ—Ä–∫–∞ (—Ç–æ–ª—å–∫–æ —Ñ–æ—Ä–º–∞—Ç)")
        print("2. –†–∞—Å—à–∏—Ä–µ–Ω–Ω–∞—è –ø—Ä–æ–≤–µ—Ä–∫–∞ (—Å –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω—ã–º–∏ —Ü–∏—Ñ—Ä–∞–º–∏)")
        print("3. –í—ã—Ö–æ–¥")
        print("-" * 50)

        choice = input("–í–∞—à –≤—ã–±–æ—Ä (1-3): ").strip()

        if choice == '1':
            validate_inn()
        elif choice == '2':
            validate_inn_with_check_digits()
        elif choice == '3':
            print("–î–æ —Å–≤–∏–¥–∞–Ω–∏—è!")
            break
        else:
            print("–ù–µ–≤–µ—Ä–Ω—ã–π –≤—ã–±–æ—Ä. –ü–æ–∂–∞–ª—É–π—Å—Ç–∞, –≤–≤–µ–¥–∏—Ç–µ 1, 2 –∏–ª–∏ 3.")

def batch_inn_validation():
    """
    –ü–∞–∫–µ—Ç–Ω–∞—è –ø—Ä–æ–≤–µ—Ä–∫–∞ –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö –ò–ù–ù
    """
    print("\n" + "=" * 50)
    print("–ü–ê–ö–ï–¢–ù–ê–Ø –ü–†–û–í–ï–†–ö–ê –ò–ù–ù")
    print("=" * 50)
    print("–í–≤–µ–¥–∏—Ç–µ –ò–ù–ù –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ (–∫–∞–∂–¥—ã–π –Ω–∞ –Ω–æ–≤–æ–π —Å—Ç—Ä–æ–∫–µ)")
    print("–î–ª—è –∑–∞–≤–µ—Ä—à–µ–Ω–∏—è –≤–≤–µ–¥–∏—Ç–µ –ø—É—Å—Ç—É—é —Å—Ç—Ä–æ–∫—É")
    print("-" * 50)

    valid_inns = []
    invalid_inns = []
    line_num = 1

    while True:
        inn = input(f"–ò–ù–ù #{line_num}: ").strip()

        if not inn:
            break

        if re.match(r'^\d{12}$', inn):
            valid_inns.append(inn)
            print(f"    –ö–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –ò–ù–ù")
        else:
            invalid_inns.append(inn)
            print(f"   –ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –ò–ù–ù (–¥–æ–ª–∂–µ–Ω –±—ã—Ç—å 12 —Ü–∏—Ñ—Ä)")

        line_num += 1

    print("\n" + "=" * 50)
    print("–†–ï–ó–£–õ–¨–¢–ê–¢–´ –ü–†–û–í–ï–†–ö–ò:")
    print("=" * 50)
    print(f" –ö–æ—Ä—Ä–µ–∫—Ç–Ω—ã—Ö –ò–ù–ù: {len(valid_inns)}")
    if valid_inns:
        print("   " + "\n   ".join(valid_inns))

    print(f"\n –ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã—Ö –ò–ù–ù: {len(invalid_inns)}")
    if invalid_inns:
        print("   " + "\n   ".join(invalid_inns))

    return valid_inns, invalid_inns

if __name__ == "__main__":
    while True:
        print("\n" + "=" * 60)
        print(" –°–ò–°–¢–ï–ú–ê –ü–†–û–í–ï–†–ö–ò –ò–ù–ù")
        print("=" * 60)
        print("1. –ü—Ä–æ–≤–µ—Ä–∏—Ç—å –æ–¥–∏–Ω –ò–ù–ù")
        print("2. –ü–∞–∫–µ—Ç–Ω–∞—è –ø—Ä–æ–≤–µ—Ä–∫–∞ –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö –ò–ù–ù")
        print("3. –í—ã—Ö–æ–¥")
        print("-" * 60)

        mode = input("–í—ã–±–µ—Ä–∏—Ç–µ —Ä–µ–∂–∏–º —Ä–∞–±–æ—Ç—ã (1-3): ").strip()

        if mode == '1':
            validate_inn_menu()
        elif mode == '2':
            batch_inn_validation()
        elif mode == '3':
            print("–î–æ —Å–≤–∏–¥–∞–Ω–∏—è!")
            break
        else:
            print(" –ù–µ–≤–µ—Ä–Ω—ã–π –≤—ã–±–æ—Ä. –ü–æ–∂–∞–ª—É–π—Å—Ç–∞, –≤–≤–µ–¥–∏—Ç–µ 1, 2 –∏–ª–∏ 3.")
```

```
import re

def validate_password():
    """
    –§—É–Ω–∫—Ü–∏—è –¥–ª—è –≤–∞–ª–∏–¥–∞—Ü–∏–∏ –ø–∞—Ä–æ–ª—è.
    –¢—Ä–µ–±–æ–≤–∞–Ω–∏—è:
    - –î–ª–∏–Ω–∞ –æ—Ç 8 –¥–æ 20 —Å–∏–º–≤–æ–ª–æ–≤
    - –•–æ—Ç—è –±—ã –æ–¥–Ω–∞ –∑–∞–≥–ª–∞–≤–Ω–∞—è –±—É–∫–≤–∞ (A-Z)
    - –•–æ—Ç—è –±—ã –æ–¥–Ω–∞ —Ü–∏—Ñ—Ä–∞ (0-9)
    –ò—Å–ø–æ–ª—å–∑—É–µ—Ç while True, try...except –∏ —Ä–µ–≥—É–ª—è—Ä–Ω—ã–µ –≤—ã—Ä–∞–∂–µ–Ω–∏—è.
    """

    print("=" * 50)
    print(" –ü–†–û–í–ï–†–ö–ê –ü–ê–†–û–õ–Ø")
    print("=" * 50)
    print("–¢—Ä–µ–±–æ–≤–∞–Ω–∏—è –∫ –ø–∞—Ä–æ–ª—é:")
    print("‚Ä¢ –î–ª–∏–Ω–∞: –æ—Ç 8 –¥–æ 20 —Å–∏–º–≤–æ–ª–æ–≤")
    print("‚Ä¢ –•–æ—Ç—è –±—ã –æ–¥–Ω–∞ –ó–ê–ì–õ–ê–í–ù–ê–Ø –±—É–∫–≤–∞")
    print("‚Ä¢ –•–æ—Ç—è –±—ã –æ–¥–Ω–∞ —Ü–∏—Ñ—Ä–∞")
    print("-" * 50)

    while True:
        try:
            password = input("–í–≤–µ–¥–∏—Ç–µ –ø–∞—Ä–æ–ª—å: ")

            if not password:
                raise ValueError("–ü–∞—Ä–æ–ª—å –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å –ø—É—Å—Ç—ã–º!")

            if len(password) < 8:
                raise ValueError("–ü–∞—Ä–æ–ª—å —Å–ª–∏—à–∫–æ–º –∫–æ—Ä–æ—Ç–∫–∏–π! –î–æ–ª–∂–µ–Ω –±—ã—Ç—å –æ—Ç 8 –¥–æ 20 —Å–∏–º–≤–æ–ª–æ–≤.")

            if len(password) > 20:
                raise ValueError("–ü–∞—Ä–æ–ª—å —Å–ª–∏—à–∫–æ–º –¥–ª–∏–Ω–Ω—ã–π! –î–æ–ª–∂–µ–Ω –±—ã—Ç—å –æ—Ç 8 –¥–æ 20 —Å–∏–º–≤–æ–ª–æ–≤.")

            if not re.search(r'[A-Z]', password):
                raise ValueError("–ü–∞—Ä–æ–ª—å –¥–æ–ª–∂–µ–Ω —Å–æ–¥–µ—Ä–∂–∞—Ç—å —Ö–æ—Ç—è –±—ã –æ–¥–Ω—É –ó–ê–ì–õ–ê–í–ù–£–Æ –±—É–∫–≤—É!")

            if not re.search(r'[0-9]', password):
                raise ValueError("–ü–∞—Ä–æ–ª—å –¥–æ–ª–∂–µ–Ω —Å–æ–¥–µ—Ä–∂–∞—Ç—å —Ö–æ—Ç—è –±—ã –æ–¥–Ω—É —Ü–∏—Ñ—Ä—É!")

            print("\n‚úÖ –ü–∞—Ä–æ–ª—å —É—Å–ø–µ—à–Ω–æ –ø—Ä–æ—à–µ–ª –≤–∞–ª–∏–¥–∞—Ü–∏—é!")
            print(f"   –î–ª–∏–Ω–∞ –ø–∞—Ä–æ–ª—è: {len(password)} —Å–∏–º–≤–æ–ª–æ–≤")
            print(f"   –ü–∞—Ä–æ–ª—å: {'‚Ä¢' * len(password)}")

            uppercase_count = sum(1 for c in password if c.isupper())
            digit_count = sum(1 for c in password if c.isdigit())

            print(f"   –ó–∞–≥–ª–∞–≤–Ω—ã—Ö –±—É–∫–≤: {uppercase_count}")
            print(f"   –¶–∏—Ñ—Ä: {digit_count}")

            return password

        except ValueError as e:
            print(f"‚ùå –û—à–∏–±–∫–∞: {e}")
            print("–ü–æ–ø—Ä–æ–±—É–π—Ç–µ —Å–Ω–æ–≤–∞...\n")

def validate_password_detailed():
    """
    –ë–æ–ª–µ–µ –ø–æ–¥—Ä–æ–±–Ω–∞—è –≤–µ—Ä—Å–∏—è —Å –æ—Ç–¥–µ–ª—å–Ω—ã–º–∏ –ø—Ä–æ–≤–µ—Ä–∫–∞–º–∏ –∫–∞–∂–¥–æ–≥–æ —Ç—Ä–µ–±–æ–≤–∞–Ω–∏—è
    """

    print("=" * 50)
    print(" –ü–†–û–í–ï–†–ö–ê –ü–ê–†–û–õ–Ø")
    print("=" * 50)

    while True:
        try:
            password = input("–í–≤–µ–¥–∏—Ç–µ –ø–∞—Ä–æ–ª—å: ")

            errors = []

            if not password:
                errors.append("–ü–∞—Ä–æ–ª—å –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å –ø—É—Å—Ç—ã–º")

            if len(password) < 8:
                errors.append(f"–°–ª–∏—à–∫–æ–º –∫–æ—Ä–æ—Ç–∫–∏–π ({len(password)} —Å–∏–º–≤–æ–ª–æ–≤). –ú–∏–Ω–∏–º—É–º 8")
            elif len(password) > 20:
                errors.append(f"–°–ª–∏—à–∫–æ–º –¥–ª–∏–Ω–Ω—ã–π ({len(password)} —Å–∏–º–≤–æ–ª–æ–≤). –ú–∞–∫—Å–∏–º—É–º 20")

            if not re.search(r'[A-Z]', password):
                errors.append("–ù–µ—Ç –∑–∞–≥–ª–∞–≤–Ω—ã—Ö –±—É–∫–≤")

            if not re.search(r'[0-9]', password):
                errors.append("–ù–µ—Ç —Ü–∏—Ñ—Ä")

            if errors:
                error_message = "\n   ‚Ä¢ ".join(["–û–±–Ω–∞—Ä—É–∂–µ–Ω—ã –ø—Ä–æ–±–ª–µ–º—ã:"] + errors)
                raise ValueError(error_message)

            print("\n" + "=" * 40)
            print("‚úÖ –ü–ê–†–û–õ–¨ –ü–†–ò–ù–Ø–¢!")
            print("=" * 40)

            print(f"   –î–ª–∏–Ω–∞: {len(password)} —Å–∏–º–≤–æ–ª–æ–≤")
            print(f"   –ó–∞–≥–ª–∞–≤–Ω—ã—Ö –±—É–∫–≤: {sum(1 for c in password if c.isupper())}")
            print(f"   –¶–∏—Ñ—Ä: {sum(1 for c in password if c.isdigit())}")
            print(f"   –°—Ç—Ä–æ—á–Ω—ã—Ö –±—É–∫–≤: {sum(1 for c in password if c.islower())}")
            print(f"   –î—Ä—É–≥–∏—Ö —Å–∏–º–≤–æ–ª–æ–≤: {sum(1 for c in password if not c.isalnum())}")

            masked = ''
            for c in password:
                if c.isupper():
                    masked += '‚¨ÜÔ∏è'
                elif c.islower():
                    masked += '‚¨áÔ∏è'
                elif c.isdigit():
                    masked += '#Ô∏è‚É£'
                else:
                    masked += 'üîπ'

            print(f"\n   –ú–∞—Å–∫–∞: {masked}")

            return password

        except ValueError as e:
            print(f"‚ùå –û—à–∏–±–∫–∞:\n   ‚Ä¢ {e}")
            print("–ü–æ–ø—Ä–æ–±—É–π—Ç–µ —Å–Ω–æ–≤–∞...\n")

def validate_password_minimal():
    """–ú–∏–Ω–∏–º–∞–ª–∏—Å—Ç–∏—á–Ω–∞—è –≤–∞–ª–∏–¥–∞—Ü–∏—è –ø–∞—Ä–æ–ª—è"""

    while True:
        password = input("–ü–∞—Ä–æ–ª—å: ")

        if len(password) < 8:
            print("‚ùå –°–ª–∏—à–∫–æ–º –∫–æ—Ä–æ—Ç–∫–∏–π (–º–∏–Ω–∏–º—É–º 8)")
            continue

        if len(password) > 20:
            print("‚ùå –°–ª–∏—à–∫–æ–º –¥–ª–∏–Ω–Ω—ã–π (–º–∞–∫—Å–∏–º—É–º 20)")
            continue

        if not re.search(r'[A-Z]', password):
            print("‚ùå –ù—É–∂–Ω–∞ –∑–∞–≥–ª–∞–≤–Ω–∞—è –±—É–∫–≤–∞")
            continue

        if not re.search(r'[0-9]', password):
            print("‚ùå –ù—É–∂–Ω–∞ —Ü–∏—Ñ—Ä–∞")
            continue

        print(f"‚úÖ –ü–∞—Ä–æ–ª—å –ø—Ä–∏–Ω—è—Ç!")
        return password

if __name__ == "__main__":
    validate_password()
```
